```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
df=yf.download('TSLA')
df.head()
```

```python
df.info()
```

```python
df.tail()
```

```python
df.columns=["close","high","low","open","volume"]
```

```python
df.head()
```

```python
df["open"].plot(figsize=(15,6))
```

```python
#Simple Moving Average
df["10 days moving average:open"]=df["open"].rolling(10).mean()
df.head(11)
```

```python
df[["open","10 days moving average:open"]].plot(xlim=["2020-01-01","2023-01-01"],figsize=(15,6))
```

```python
df["30 days moving average:open"]=df["open"].rolling(30).mean()
df["50 days moving average:open"]=df["open"].rolling(50).mean()
```

```python
df[["10 days moving average:open","30 days moving average:open","50 days moving average:open"]].plot(xlim=["2020-01-01","2023-01-01"],figsize=(15,6))
```

```python
#Cumilative moving average
#Expanding
df["open"].expanding().mean().plot(figsize=(15,6))
```

```python
df.head()
```

```python
df["CMA:open"]=df["open"].expanding().mean()
```

```python
df.head()
```

```python
df[["open","CMA:open"]].plot(figsize=(15,6))
```

```python
#Exponential Moving Average -EWMA
```

```python
#with smoothing factor 0.1 i.e alpha
df["EMA_0.1"]=df["open"].ewm(alpha=0.1,adjust=False).mean()
```

```python
df[["open","EMA_0.1"]].plot(figsize=(15,6))
```

```python
#with change in smoothing hyperparameter
df["EMA_0.3"]=df["open"].ewm(alpha=0.3,adjust=False).mean()
```

```python
df[["open","EMA_0.1","EMA_0.3"]].plot(xlim=["2020-01-01","2022-01-01"],figsize=(15,6))
```

```python
#using span hyperparameter
df["open_ewa5days"]=df["open"].ewm(span=5).mean()
```

```python
df[["open","open_ewa5days","EMA_0.1"]].plot(xlim=["2020-01-01","2022-01-01"],figsize=(15,6))
```
