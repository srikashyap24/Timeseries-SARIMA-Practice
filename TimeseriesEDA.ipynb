#### Time Resampling 

```python
df.head()
```

```python
# Resampling in Exploratory Data Analysis (EDA) of time series data refers to changing the frequency of observations by either up-sampling (increasing frequency) or down-sampling (decreasing frequency). 
#This is useful for understanding trends, seasonality, and missing values in time series data.

# Types of Resampling:

# Down-sampling (Reducing Frequency)
# Aggregates data to a lower frequency (e.g., hourly data  daily data).
# Example: Converting minute-wise stock prices to daily closing prices.
# Methods: mean(), sum(), min(), max()
# Use case: Reducing noise and making long-term trends clearer.

# Up-sampling (Increasing Frequency)
# Increases data points by filling missing timestamps (e.g., daily  hourly).
# Interpolation methods: ffill() (forward fill), bfill() (backward fill), linear interpolation.
# Use case: Making data compatible with higher-frequency datasets or improving model accuracy.
```

```python
# df.resample(rule='A') is a down-sampling operation in pandas that resamples the time series data annually (end of the year).

# Explanation of A (Annual Frequency)
# 'A' stands for year-end frequency (Annual).
# It groups data by the last day of each year and applies an aggregation function (e.g., mean(), sum(), etc.).
# Equivalent to resampling at December 31st of every year.

# Other Variations of Annual Resampling:

# 'A'	Year-end (December 31st)
# 'A-JUN'	Fiscal year ending in June
# 'A-MAR'	Fiscal year ending in March
# 'AS'	Year-start (January 1st)
df.resample(rule='A')
```

```python
df.resample(rule='A').max()
```

```python
df.resample(rule='A').min()["open"].plot(figsize=(12,4))
```

```python
df.resample(rule='A').max()["close"].plot(figsize=(12,4))
```

```python
#Quaterly start resampling
df.resample(rule="QS").min()["high"].plot(figsize=(12,4))
```

```python
#Business End Frequency 
df.resample(rule='BA').max()
```

```python
#Business quaterly start frequency 
df.resample(rule='BQS').max()
```

```python
df.resample(rule='BQS').mean()["open"].plot(kind='bar')
```

```python
#Monthly
df["open"].resample(rule='ME').mean().plot(figsize=(15,6),kind="bar")
```
